<h4>Import Libraries</h4>

```python
import pandas as pd
import numpy as np
```

<h4>Import Dataset</h4>

```python
dataset=pd.read_csv("F:/DB/petrol_consumption.csv")
```

<h4>Seprate Independent and Dependentent Columns from Dataset</h4>

```python
X=dataset.iloc[:,0:4].values
y=dataset.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

<h4>Fitting Decison tree Regression to the Training set</h4>

```python

from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor()
reg.fit(X_train,y_train)

pred=reg.predict(X_test)
```

<h4>Model Evalution</h4>

```python
from sklearn import metrics
print("Mean Square Error",metrics.mean_squared_error(y_test,pred))
print("Roor mean Square Error",np.sqrt(metrics.mean_squared_error(y_test,pred)))
```

<h4>Display tree</h4>

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
dot_data = StringIO()
export_graphviz(reg, out_file=dot_data, 
                
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
